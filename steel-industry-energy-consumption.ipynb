```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

def encode_categorical(df, nominal_columns, ordinal_columns, ordinal_mappings):
    df_encoded = df.copy()

    # Ordinal Encoding
    for col in ordinal_columns:
        if col in ordinal_mappings:
            df_encoded[col] = df_encoded[col].map(ordinal_mappings[col])
        else:
            # If no specific mapping provided, use label encoding (not always ideal for ordinal)
            le = LabelEncoder()
            df_encoded[col] = le.fit_transform(df_encoded[col])

    # One-Hot Encoding
    df_encoded = pd.get_dummies(df_encoded, columns=nominal_columns, drop_first=True)

    return df_encoded

# Specify which columns are nominal and which are ordinal
nominal_columns = ['WeekStatus','Day_of_week']
ordinal_columns = ['Load_Type']

# Define mappings for ordinal columns
ordinal_mappings = {
    'Load_Type': {'Light_Load': 1, 'Medium_Load': 2, 'Maximum_Load': 3}
}

# Encode the DataFrame
df_encoded = encode_categorical(df, nominal_columns, ordinal_columns, ordinal_mappings)

print(df_encoded)
```

# Exploratary Data Analysis (EDA)

```python
df.describe()
```

```python
import pandas as pd

# Assuming df is your DataFrame
summary_stats = df.describe()

# Function to calculate coefficient of variation
def coefficient_of_variation(mean, std):
    if mean != 0:
        return std / mean
    else:
        return float('inf')

# Evaluating each attribute
for column in summary_stats.columns:
    mean = summary_stats.loc['mean', column]
    std = summary_stats.loc['std', column]
    min_value = summary_stats.loc['min', column]
    q1 = summary_stats.loc['25%', column]
    median = summary_stats.loc['50%', column]
    q3 = summary_stats.loc['75%', column]
    max_value = summary_stats.loc['max', column]
    
    # Calculate CV
    cv = coefficient_of_variation(mean, std)
    
    # Evaluate variability
    if cv > 1:
        variability_evaluation = "High Variability"
    elif cv > 0.5:
        variability_evaluation = "Moderate Variability"
    else:
        variability_evaluation = "Low Variability"
    
    # Evaluate skewness
    if mean > 1.5 * median:
        skewness_evaluation = "Right-skewed distribution"
    elif median > 1.5 * mean:
        skewness_evaluation = "Left-skewed distribution"
    else:
        skewness_evaluation = "Fairly symmetric distribution"
    
    # Document the findings
    print(f"### {column}")
    print(f"**Evaluation**:")
    print(f"- **Mean**: {mean:.2f}")
    print(f"- **Std**: {std:.2f}")
    print(f"- **Min**: {min_value:.2f}")
    print(f"- **25%**: {q1:.2f}")
    print(f"- **50% (Median)**: {median:.2f}")
    print(f"- **75%**: {q3:.2f}")
    print(f"- **Max**: {max_value:.2f}")
    print(f"- **Coefficient of Variation (CV)**: {cv:.2f}")
    print(f"- **Variability**: {variability_evaluation}")
    print(f"- **Skewness**: {skewness_evaluation}")
    print(f"**Conclusion**: This attribute has {variability_evaluation.lower()} and is {skewness_evaluation.lower()}.")
    if cv > 1 or mean > 1.5 * median:
        print("Transformation or normalization might be necessary.")
    print("\n")
```

# Check the Multicollinearity

# Variance Inflation Factor (VIF) = 1/(1 - R^2) 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Assuming df_encoded is already loaded and contains the necessary columns
# For VIF calculation, ensure df_encoded only includes numerical columns
df_numerical = df_encoded.select_dtypes(include=['float64', 'int64'])

# Add constant term for intercept
df_numerical_with_const = add_constant(df_numerical)

# Calculate VIF for each variable
vif_data = pd.DataFrame()
vif_data["Variable"] = df_numerical_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(df_numerical_with_const.values, i) for i in range(df_numerical_with_const.shape[1])]

# Print VIF DataFrame
print(vif_data)

plt.show()
```

# Interpretation:
High VIF Values: A VIF value greater than 5 or 10 typically indicates high multicollinearity.

* Usage_kWh (50.77)
* CO2(tCO2) (43.33)
* Leading_Current_Reactive_Power_kVarh (10.30)
* Leading_Current_Power_Factor (12.33)
* WeekStatus_Weekend (inf)
* Day_of_week_Saturday (inf)
* Day_of_week_Sunday (inf)

Infinite VIF: The inf value for WeekStatus_Weekend, Day_of_week_Saturday, and Day_of_week_Sunday indicates perfect multicollinearity.
