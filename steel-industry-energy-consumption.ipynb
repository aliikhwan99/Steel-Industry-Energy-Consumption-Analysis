```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

def encode_categorical(df, nominal_columns, ordinal_columns, ordinal_mappings):
    df_encoded = df.copy()

    # Ordinal Encoding
    for col in ordinal_columns:
        if col in ordinal_mappings:
            df_encoded[col] = df_encoded[col].map(ordinal_mappings[col])
        else:
            # If no specific mapping provided, use label encoding (not always ideal for ordinal)
            le = LabelEncoder()
            df_encoded[col] = le.fit_transform(df_encoded[col])

    # One-Hot Encoding
    df_encoded = pd.get_dummies(df_encoded, columns=nominal_columns, drop_first=True)

    return df_encoded

# Specify which columns are nominal and which are ordinal
nominal_columns = ['WeekStatus','Day_of_week']
ordinal_columns = ['Load_Type']

# Define mappings for ordinal columns
ordinal_mappings = {
    'Load_Type': {'Light_Load': 1, 'Medium_Load': 2, 'Maximum_Load': 3}
}

# Encode the DataFrame
df_encoded = encode_categorical(df, nominal_columns, ordinal_columns, ordinal_mappings)

print(df_encoded)
```

```python
import pandas as pd

# Assuming df is your DataFrame
summary_stats = df.describe()

# Function to calculate coefficient of variation
def coefficient_of_variation(mean, std):
    if mean != 0:
        return std / mean
    else:
        return float('inf')

# Evaluating each attribute
for column in summary_stats.columns:
    mean = summary_stats.loc['mean', column]
    std = summary_stats.loc['std', column]
    min_value = summary_stats.loc['min', column]
    q1 = summary_stats.loc['25%', column]
    median = summary_stats.loc['50%', column]
    q3 = summary_stats.loc['75%', column]
    max_value = summary_stats.loc['max', column]
    
    # Calculate CV
    cv = coefficient_of_variation(mean, std)
    
    # Evaluate variability
    if cv > 1:
        variability_evaluation = "High Variability"
    elif cv > 0.5:
        variability_evaluation = "Moderate Variability"
    else:
        variability_evaluation = "Low Variability"
    
    # Evaluate skewness
    if mean > 1.5 * median:
        skewness_evaluation = "Right-skewed distribution"
    elif median > 1.5 * mean:
        skewness_evaluation = "Left-skewed distribution"
    else:
        skewness_evaluation = "Fairly symmetric distribution"
    
    # Document the findings
    print(f"### {column}")
    print(f"**Evaluation**:")
    print(f"- **Mean**: {mean:.2f}")
    print(f"- **Std**: {std:.2f}")
    print(f"- **Min**: {min_value:.2f}")
    print(f"- **25%**: {q1:.2f}")
    print(f"- **50% (Median)**: {median:.2f}")
    print(f"- **75%**: {q3:.2f}")
    print(f"- **Max**: {max_value:.2f}")
    print(f"- **Coefficient of Variation (CV)**: {cv:.2f}")
    print(f"- **Variability**: {variability_evaluation}")
    print(f"- **Skewness**: {skewness_evaluation}")
    print(f"**Conclusion**: This attribute has {variability_evaluation.lower()} and is {skewness_evaluation.lower()}.")
    if cv > 1 or mean > 1.5 * median:
        print("Transformation or normalization might be necessary.")
    print("\n")
```

```python
import pandas as pd
from scipy.stats import spearmanr, pointbiserialr, chi2_contingency, kendalltau
import numpy as np

# Select columns based on their types
continuous_vars = ['Usage_kWh', 'Lagging_Current_Reactive.Power_kVarh', 'Leading_Current_Reactive_Power_kVarh', 
                   'CO2(tCO2)', 'Lagging_Current_Power_Factor', 'Leading_Current_Power_Factor', 
                   'Log_Lagging_Current_Power_Factor', 'Log_Leading_Current_Power_Factor', 'BoxCox_Usage_kWh', 
                   'BoxCox_Lagging_Current_Reactive_Power_kVarh', 'BoxCox_Leading_Current_Reactive_Power_kVarh']

binary_vars = ['WeekStatus']
categorical_vars = ['Day_of_week', 'Load_Type']

# 1. Spearman Correlation for continuous variables
spearman_corr = df[continuous_vars].corr(method='spearman')
print("Spearman Correlation (Continuous Variables):")
print(spearman_corr)

# 2. Rank-Biserial Correlation for binary-continuous pairs
for binary_col in binary_vars:
    for continuous_col in continuous_vars:
        correlation, p_value = spearmanr(df[binary_col], df[continuous_col])
        print(f"Spearman correlation between {binary_col} and {continuous_col}: {correlation}, p-value: {p_value}")

# 3. CramÃ©r's V for categorical-categorical variables
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

for cat1 in categorical_vars:
    for cat2 in categorical_vars:
        if cat1 != cat2:
            cramer_v_value = cramers_v(df[cat1], df[cat2])
            print(f"CramÃ©r's V between {cat1} and {cat2}: {cramer_v_value}")

# 4. Kendall's Tau for continuous-categorical pairs
for cat_col in categorical_vars:
    for continuous_col in continuous_vars:
        tau, p_value = kendalltau(df[cat_col], df[continuous_col])
        print(f"Kendallâ€™s Tau between {cat_col} and {continuous_col}: {tau}, p-value: {p_value}")
```

# Categorical-Categorical Correlation (CramÃ©r's V) - for Nominal

```python
# Define a function to calculate CramÃ©r's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

# Categorical variables
categorical_vars = ['Day_of_week','Load_Type']

# Store CramÃ©r's V values in a DataFrame
cramers_v_matrix = pd.DataFrame(index=categorical_vars, columns=categorical_vars)

# Calculate CramÃ©r's V for all pairs of categorical variables
for cat1 in categorical_vars:
    for cat2 in categorical_vars:
        if cat1 != cat2:
            cramers_v_matrix.loc[cat1, cat2] = cramers_v(df[cat1], df[cat2])

# Plot CramÃ©r's V heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(cramers_v_matrix.astype(float), annot=True, cmap='coolwarm', center=0, vmin=0, vmax=1, linewidths=0.5)
plt.title("CramÃ©r's V for Categorical Variables")
plt.show()
```

# Use Kruskal-Wallis ANOVA (for data with unequal variances & non-parametric)

```python
import scipy.stats as stats

# Kruskal-Wallis test for 'Usage_kWh' across 'Load_Type'
# Group the data by 'Load_Type'
groups = [df[df['Load_Type'] == load]['Usage_kWh'] for load in df['Load_Type'].unique()]

# Perform Kruskal-Wallis ANOVA
stat, p = stats.kruskal(*groups)

print(f"Kruskal-Wallis Test: stat={stat:.3f}, p={p:.3f}")
```

# Check the multicollinearity after apply regression model

Multicollinearity:
This occurs when two or more **independent variables** in a regression model are highly correlated, making it difficult to separate out the individual effects of each variable.

# Requirement before applying VIF

* Focus on Independent Variables: Include only the independent variables (predictors) in the VIF analysis.
* Numerical Data: Ensure variables are numerical, converting categorical variables into dummy variables.
* Clean Data: Make sure the data is clean, with no missing values or outliers.
* Multiple Variables: VIF is meaningful only with multiple independent variables.

# Variance Inflation Factor (VIF) = 1/(1 - R^2) 

VIF Definition:

VIF measures how much the variance of a regression coefficient is inflated due to multicollinearity. It helps identify whether an independent variable is highly correlated with other independent variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# Assuming df_encoded is already loaded and contains the necessary columns
# For VIF calculation, ensure df_encoded only includes numerical columns
df_numerical = df_encoded.select_dtypes(include=['float64', 'int64'])

# Add constant term for intercept
df_numerical_with_const = add_constant(df_numerical)

# Calculate VIF for each variable
vif_data = pd.DataFrame()
vif_data["Variable"] = df_numerical_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(df_numerical_with_const.values, i) for i in range(df_numerical_with_const.shape[1])]

# Print VIF DataFrame
print(vif_data)

plt.show()
```

# Interpretation:

Understand The **constat value of the VIF** when VIF intercept is not relevant just focus on VIFs of the independent variables only.

**Intercept (Constant):** Represents the baseline level of the dependent variable when all independent variables are zero. Itâ€™s a parameter in the regression equation, not related to multicollinearity.
